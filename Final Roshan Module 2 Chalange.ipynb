```python
# Import the pathlib and csv library
from pathlib import Path
import csv
```

```python
# Set the file path
csvpath = Path('../Resources/budget_data.csv')
```

```python
#open input path as file object
with open(csvpath, "r") as csvfile:
    
    #Initialize variables
    count = 1
    minimum = [0, '']
    maximum = [0, '']
    all_change_pnl = []
    
    
    #pass in csv file to csv.reader() function
    csvreader = csv.reader(csvfile, delimiter=',')
    
    #set header as first row
    header = next(csvreader)
    
    #set line_1 as one afer the header 
    line_1 = next(csvreader)
    
    #initialize total profit/loss 
    total_pnl_amount = int(line_1[1])
    
    #initialize change in profit/loss for each month variable
    total_change_pnl = 0
    
    #Looping thrugh each row of data after the header
    
    for row in csvreader:
        change_pnl = int(row[1]) - int(line_1[1])

    
    #The newt total amount of profit/losses over entire period
        total_pnl_amount += int(row[1]) 
        
    #total number of months included in data set
        count += 1
        
    # Setting New Line 1 inside the for loop      
        line_1 = row
        
    #append change to list
        all_change_pnl.append(change_pnl)
        
    #The average of the changes in Profit/Losses over the entire period
        average_change_pnl = (sum(all_change_pnl)/len(all_change_pnl))
        
    #logic to determine mminimum and maximum
    
        if minimum == 0:
            minimum[0] = change_pnl
            minimum[1] = row[0]
        elif change_pnl < minimum[0]:
            minimum[0] = change_pnl
            minimum[1] = row[0]
        elif change_pnl > maximum[0]:
            maximum[0] = change_pnl
            maximum[1] = row[0]
            
            
   #The result analysis     
        
    print("Financial Analysis")
    print("----------------------------")
 
    #question 1
    print(f"Total Months:  {count}")
     
    #question 2
    print(f"Total:  ${total_pnl_amount}")
     
    #Question 3
    print(f"Average change:  ${average_change_pnl}")
    
    #question 4 
    print(f"Greates Increase in Profits: ${maximum}")
    
    #Question 5
    print(f"Greatest Decrease in Profits: ${minimum}")
    
    print("------")
    
# Set the output file path
output_path = Path('output.txt')

#Open the output path as a file object
with open(output_path, 'w') as file:
          file.write((f"Total Months:  {count}"))
          file.write((f"Total:  ${total_pnl_amount}"))
          file.write((f"Average change:  ${average_change_pnl}"))
          file.write((f"Greates Increase in Profits: ${maximum}"))
          file.write((f"Greatest Decrease in Profits: ${minimum}"))
  
```
